```python
#import graphical functionality -ABQ
import matplotlib.pyplot as plt #type: ignore

#import system directories and file navigation and management. -ABQ
import os

#here we import two variables from filePathing and assign them nicknames, filePathing is not a library or module, but an archive, filePathing.py, where inside it resides global string variables that contains the path to the relevant directories and files. this way, management of file paths of different devices is easier and reduces a bit of code clutter -ABQ
from filePathing import pathToDataDirForThisDevice as pathToData, pathToGraphsForThisDevice as pathToGraphs

#import array sweep functions from arrayFunctions.py. we also give them some shorter nicknames -ABQ
from arrayFunctions import  findRowWithMMsClosestTo0 as findMMsZero, findRowWithField0 as findFieldZero

#this gets all the files names inside the data directory -ABQ
getListOfFiles = [os.path.join(pathToData, singleDataFile) for singleDataFile in os.listdir(pathToData) if os.path.isfile(os.path.join(pathToData, singleDataFile))] #this lists all files inside a directory -ABQ

#this is function returns half the sum of two modules. there probably is a function that does this in the python math library, but i just find easier to do this standalone function instead of searching it and importing from the library -ABQ
def getAbsMean(targetNumber1, targetNumber2):
  absMeanIs = (abs(targetNumber1)+abs(targetNumber2))/2
  return absMeanIs

#here with declare some arrays that we gonna work with later, mostly appending single row arrays to them -ABQ
arrayOfSampleNames = []
arrayOfReads = []
arrayOfAvgOfCoerciveFields = []
arrayOfMrs = []

#loop for sweeping all files and reading their content. in this case, it has has a 2 step increment cycle -ABQ
for counter in range(0, len(getListOfFiles), 2):
  counterFor00Degrees = counter
  counterFor90Degrees = (counter + 1) if counter + 1 < len(getListOfFiles) else None

  #dataset 1, in this case, we work with the 00° readings -ABQ
  with open(getListOfFiles[counterFor00Degrees], "r") as file1:
    #this reads the sample names as integers so they may go into the graphs or whatever we want to do with them. notice that in this case, we only do it once inside the loop, using the 00° file -ABQ
    readSampleName = int(os.path.splitext(os.path.basename(getListOfFiles[counterFor00Degrees]))[0].split("_")[0])
    associateSampleNameWithArrayRow = [readSampleName]
    arrayOfSampleNames.append(associateSampleNameWithArrayRow)

    #here we read all lines in the file to later store them in a array -ABQ
    lineReadingsForDataset1 = file1.readlines()

    #we declare the array to be appended with the lines -ABQ
    arrayForDataset1 = []
    
    #and now this loop will append all lines in the file to the array -ABQ
    for line in lineReadingsForDataset1:
      singleLineForDataset1 = line.split() 
      associateFileLineWithArrayRowForDataset1 = [float(reading1) for reading1 in singleLineForDataset1]
      arrayForDataset1.append(associateFileLineWithArrayRowForDataset1)

    #here we find M/Ms closest to zero, which give us the corcitive field -ABQ
    minMMs00Degrees1stHalf = findMMsZero(arrayForDataset1[:(len(arrayForDataset1))//2])
    minMMs00Degrees2ndHalf = findMMsZero(arrayForDataset1[(len(arrayForDataset1))//2:])
    associateAvgCoerciveFieldWithArrayForDataset1 = [getAbsMean(minMMs00Degrees1stHalf[0], minMMs00Degrees2ndHalf[0])]

    #here we find Mr/Ms, which by definition is when the field is zero -ABQ
    fieldZero00Degrees1stHalf = findFieldZero(arrayForDataset1[:(len(arrayForDataset1))//2])
    fieldZero00Degrees2ndHalf = findFieldZero(arrayForDataset1[(len(arrayForDataset1))//2:])
    associateAvgField0WithArrayForDataset1 = [getAbsMean(fieldZero00Degrees1stHalf[1], fieldZero00Degrees2ndHalf[1])]

  #dataset 2, in this case, we work with the 90° readings -ABQ
  with open(getListOfFiles[counterFor90Degrees], "r") as file2:
    #here we read all lines in the file to later store them in a array -ABQ
    lineReadingsForDataset2 = file2.readlines()

    #we declare the array to be appended with the lines -ABQ
    arrayForDataset2 = []
    
    #and now this loop will append all lines in the file to the array -ABQ
    #notice that we gonna use different variable names for the different loops for the 00° and 90° open file methods, but it doesn't make a differece as is inside the loops and doesn't bleed outside it -ABQ
    for line in lineReadingsForDataset2:
      singleLineInDataset2 = line.split()
      associateFileLineWithArrayRowForDataset2 = [float(reading2) for reading2 in singleLineInDataset2]
      arrayForDataset2.append(associateFileLineWithArrayRowForDataset2)

    #here we find M/Ms closest to zero -ABQ
    minMMs90Degrees1stHalf = findMMsZero(arrayForDataset2[:(len(arrayForDataset2))//2])
    minMMs90Degrees2ndHalf = findMMsZero(arrayForDataset2[(len(arrayForDataset2))//2:])
    associateAvgCoerciveFieldWithArrayForDataset2 = [getAbsMean(minMMs90Degrees1stHalf[0], minMMs90Degrees2ndHalf[0])]

    #here we find Mr/Ms, which by definition is when the field is zero -ABQ
    fieldZero90Degrees1stHalf = findFieldZero(arrayForDataset2[:(len(arrayForDataset2))//2])
    fieldZero90Degrees2ndHalf = findFieldZero(arrayForDataset2[(len(arrayForDataset2))//2:])
    associateAvgField0WithArrayForDataset2 = [getAbsMean(fieldZero90Degrees1stHalf[1], fieldZero90Degrees2ndHalf[1])]

  #outsite "open" method array construction, still inside sweep of loop -ABQ
  #this builds up the array of coercive fields using the means of the absulute M/Ms minimums from the first and second half of the datasets -ABQ
  findMeanOfCoerciveFields = getAbsMean(associateAvgCoerciveFieldWithArrayForDataset1[0], associateAvgCoerciveFieldWithArrayForDataset2[0])
  associateAvgOfCoerciveFieldsWithArrayRow = [findMeanOfCoerciveFields]
  arrayOfAvgOfCoerciveFields.append(associateAvgOfCoerciveFieldsWithArrayRow)

  #this builds up the array of Mr/Ms using the means of the absulute minimums fields from the first and second half of the datasets -ABQ
  findMeanOfMr = getAbsMean(associateAvgField0WithArrayForDataset1[0], associateAvgField0WithArrayForDataset2[0])
  associateAvgOfCoerciveFields0WithArrayRow = [findMeanOfMr]
  arrayOfMrs.append(associateAvgOfCoerciveFields0WithArrayRow)
  
  #this print us the actual numbers for the coercive fields and Mr/Ms for the respective samples -ABQ
  #the round function is selfexmplanatory, the method 2 is for the 2 numbers after the decimal case, it may assume any value we want or be removed altogether. -ABQ
  print(f"Para {readSampleName}nm, Hc é {round(associateAvgOfCoerciveFieldsWithArrayRow[0], 2)} Oe e Mr/Ms é {round(associateAvgOfCoerciveFields0WithArrayRow[0], 2)} arb. units.")


#build axes from array -ABQ
xAxisReadingsForDataset1 = [row[0] for row in arrayOfSampleNames]
yAxisReadingsForDataset1 = [row[0] for row in arrayOfAvgOfCoerciveFields]
yAxisTwinReadingsForDataset1 = [row[0] for row in arrayOfMrs]


#plot graph -ABQ
#firstly, since we want twin y axis, we use the subplot function -ABQ
fig, ax1 = plt.subplots()

#the line declarations serves so we can add nice legends to the graph later -ABQ
line1, = ax1.plot(xAxisReadingsForDataset1, yAxisReadingsForDataset1, "--o", color="r", markersize=5, label="Hc")
ax1.set_xlabel("Espessura (nm)")
ax1.set_xlim(-25, max(xAxisReadingsForDataset1)+25)
ax1.set_ylabel("H (Oe)")
ax1.tick_params(axis="y")
ax1.set_ylim(-25, 300)

ax2 = ax1.twinx()
line2, = ax2.plot(xAxisReadingsForDataset1, yAxisTwinReadingsForDataset1, "--^", color="g", markersize=5, label="Mr/Ms (arb. unit)")
ax2.set_ylabel("M/Ms (arb. unit)")
ax2.tick_params(axis="y")
ax2.set_ylim(0, 1+0.025)

lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc="upper right", frameon=False)

ax3 = ax1.twiny()
ax3.set_xlim(ax1.get_xlim())
ax3.tick_params(axis='x', which='both', bottom=False, top=True, labelbottom=False, labeltop=False)

#this merely saves the generated graph to a png in the graphs directory -ABQ
plt.savefig(os.path.join(pathToGraphs, f"graph_ThicknessEvolution.png"))

#boilerplate
plt.show()
```
